# Q5 Batsmen came to crease but never got a chance to face a ball

```python
ides=df['match_id'].unique()
```

```python
players_not_get_chance=[]
grp=df.groupby('match_id')
for id in ides:
    df_id=grp.get_group(id)
    batsman=df_id['batsman']
    non_stricker=df_id['non_striker']
    player=np.setdiff1d(non_stricker,batsman)
    if len(player)>0:
        players_not_get_chance.append(player)
```

```python
players_not_get_chance
```
